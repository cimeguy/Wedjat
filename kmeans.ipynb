```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Example data: 2D points
np.random.seed(0)
points = np.random.rand(100, 2)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(points)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plotting the clusters and centroids
plt.figure(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("KMeans Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

```

```python
distances = kmeans.transform(points)
```

```python
print(distances[0], labels[0])# distance to each centroid, label of the closest centroid
all_distances_c0 = distances[:,0] # distance to the second closest centroid
all_distances_c1 = distances[:,1] # distance to the second closest centroid
all_distances_c2 = distances[:,2] # distance to the second closest centroid
```

```python
kmeans.cluster_centers_
```

```python
newpoint = np.array([[0.6 , 0.6 ]])
label = kmeans.predict(newpoint)[0]
new_distances = kmeans.transform(newpoint)
print(new_distances[0], label)
score_each_c = [1,-1,1]
d = new_distances[0][label]# distance to the closest centroid

print(d)
print(all_distances_c2)
# more closer, more bigger the score is 
similar_p2c= sum(all_distances_c2 > d ) / len(all_distances_c2) # percentage of points closer to the closest centroid than the new point
score = similar_p2c * score_each_c[label]
print(similar_p2c,score)
```
