# 准备数据

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
# 生成数据
X, y = make_blobs(n_samples=300, centers=4, random_state=0)
```

```python
import auto
```

```python
print(X.shape,y.shape)
print(X)
```

# SVC

```python


# 定义核函数和谱聚类对象
kernel = 'rbf'
gamma = 0.5
n_clusters = 6
clustering = SpectralClustering(n_clusters=n_clusters, affinity=kernel, gamma=gamma)

# 聚类并可视化结果
y_pred = clustering.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis')
plt.show()
```

# 绘制y_pred和y_true的频率分布直方图

```python
import scienceplots
plt.style.use(['science','ieee'])
# 绘制 y_pred 和 y 的频率分布直方图
fig, ax = plt.subplots(ncols=2, figsize=(10, 5))

bins = np.arange(-0.5, n_clusters + 0.5, 1)

ax[0].hist(y_pred, bins=bins, rwidth=0.7)
ax[0].set_xticks(range(n_clusters))
ax[0].set_xlabel('Cluster Label')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Predicted Labels')

ax[1].hist(y, bins=bins, rwidth=0.8)
ax[1].set_xticks(range(n_clusters))
ax[1].set_xlabel('Cluster Label')
ax[1].set_ylabel('Frequency')
ax[1].set_title('True Labels')

plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score

# y_pred为聚类结果，y为真实标签
print(y_pred,y)
purity = adjusted_mutual_info_score(y, y_pred)
print(adjusted_rand_score(y, y_pred))
print("Purity:", purity)
```

# 贝叶斯搜索

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_mutual_info_score
from sklearn.cluster import SpectralClustering
from skopt import BayesSearchCV
from skopt.space import Real, Integer
from skopt.utils import use_named_args
from functools import partial

# 创建模拟数据
X, y = make_blobs(n_samples=1000, centers=3, random_state=42)

# 定义评分函数
def adjusted_mutual_info_score_wrapper(model, X, y_true):
    labels = model.fit_predict(X)
    return adjusted_mutual_info_score(y_true, labels)

# 定义参数搜索空间
search_space = {
    'gamma': Real(0.1, 5, prior='log-uniform'),
    'n_clusters': Integer(2, 10)
}

# 初始化贝叶斯优化搜索对象
bayes_search = BayesSearchCV(
    estimator=SpectralClustering(n_jobs=-1, affinity='rbf'),
    search_spaces=search_space,
    n_iter=5,
    scoring=adjusted_mutual_info_score_wrapper,
    n_jobs=-1,
    cv=5
)

# 进行参数搜索
result = bayes_search.fit(X, y)

# 输出最优参数和评分
print("Best gamma:", result.best_params_['gamma'])
print("Best n_clusters:", result.best_params_['n_clusters'])
print("Best score:", -1.0 * result.best_score_)
```

## 得到部分score以便绘图

```python
import numpy as np
from itertools import product
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_mutual_info_score
from sklearn.cluster import SpectralClustering
from skopt import BayesSearchCV
from skopt.space import Real, Integer
from skopt.utils import use_named_args
from functools import partial
from skopt import plots


# 绘制参数搜索过程中不同参数组合的得分情况
gamma_range = np.logspace(-2, 2, 10)# 0.01～100  10个数
n_clusters_range = np.arange(2, 12) # 2～11  10个数
param_grid = list(product(gamma_range, n_clusters_range))

scores = []
scores_dict ={}
scores_array = np.zeros((10,10))
for i in range(len(gamma_range)):
    this_gamma = gamma_range[i]
    for j in range(len(n_clusters_range)):
        this_n_clusters = n_clusters_range[j]
        model = SpectralClustering(n_clusters=this_n_clusters, gamma=this_gamma, n_jobs=-1, affinity='rbf')
        score = adjusted_mutual_info_score(y, model.fit_predict(X))
        print(this_gamma,this_n_clusters, ": ",score)
        scores_array[i][j] = - score
        scores.append(score)

# 这个较快
# for gamma, n_clusters in param_grid:
#     model = SpectralClustering(n_clusters=n_clusters, gamma=gamma, n_jobs=-1, affinity='rbf')
#     score = adjusted_mutual_info_score(y, model.fit_predict(X))
#     print(score)
#     scores_dict[(gamma, n_clusters)] = score
#     scores.append(score)
# scores = np.array(scores)
# scores = -1.0 * scores # 统一变成负的


```
